The following cell creates a triangular mesh of size num_points:

```python
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import numpy as np

# Define the number of points
num_points = 64

# Generate coordinates
x_coords = np.linspace(0, 100, int(np.sqrt(num_points)) + 1)
y_coords = np.linspace(0, 100, int(np.sqrt(num_points)) + 1)

# Create a meshgrid
x_coords, y_coords = np.meshgrid(x_coords, y_coords)
x_coords, y_coords = x_coords.flatten(), y_coords.flatten()

# Select a subset of points to ensure even distribution
indices = np.linspace(0, len(x_coords) - 1, num_points, dtype=int)
x_coords = x_coords[indices]
y_coords = y_coords[indices]

# Create a triangulation object
triangulation = tri.Triangulation(x_coords, y_coords)

# Generate color data based on a combination of x and y coordinates
max_value = max(max(x_coords), max(y_coords))
min_value = min(min(x_coords), min(y_coords))
color_data = (0.1+((np.array(x_coords) + np.array(y_coords)) / 2 - min_value) / (max_value - min_value))*100

# Plot the mesh with color
plt.figure()
plt.tripcolor(triangulation, color_data, cmap='viridis')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
#plt.title('Triangular Mesh with Diagonal Color Gradient')
plt.show()
#plt.savefig("Mesh_64.pdf")
```

```python
"""
Create a graph with the nodes as the points in the mesh and the edges as the triangulation lines. 
The weights between the edges are the difference in color (or property) betweent the nodes of that edge.
"""

G=nx.Graph()
for i in range(num_points):
    G.add_node(i)
    
for x in triangulation.edges :
    G.add_edge(x[0],x[1])
    
for u, v in G.edges():
    G[u][v]['weight'] = abs(color_data[u]-color_data[v])
```

```python
#get the position of all the nodes, this will be important to 
pos={}
for i in range(len(G.nodes())):
    pos[i]= (x_coords[i],y_coords[i])
```

```python



def graphpart(G):
    """
    This function takes as input a graph G and divides it into two subgraphs and returns the subgraphs.

    """

    nodes=[x for x in G.nodes()]  #set of all nodes , note the they might not be sequential as the incoming
                                  #graphs themselves can be subgraphs and hence can have node and edge names 
                                    #that are not sequential
    nodedict={}             
    for i in range(len(G.nodes())):    
        nodedict[i]=nodes[i]

    #get the weighted degrees of the nodes
    weighted_degrees = dict(G.degree(weight='weight'))
    #get the adjacency matrix of the graph, A_ij in the modularity expression
    adj_matrix = nx.to_numpy_array(G)
    #sum of weighted degree of all nodes which is 2m in the modularity expression
    sum_of_degrees=sum(weighted_degrees)
    #calculate the second term of the modularity expression , (k_i*k_j/2m)
    second_term=np.zeros((len(G.nodes()), len(G.nodes())))
    for i in range(len(G.nodes())) :
        for j in range(len(G.nodes())) :
            second_term[i][j]= (weighted_degrees[nodedict[i]]*weighted_degrees[nodedict[j]])/sum_of_degrees
    #finally get the modularity matrix
    modularity=adj_matrix-second_term 
    #use the modularity matrix in the hamiltsolver function to get the solution
    a,SS, SS1=hamiltsolver(-modularity,'statevector_simulator')
    
    
    SS_resolved=[nodedict[i] for i in SS]
    SS1_resolved=[nodedict[i] for i in SS1]
    return G.subgraph(SS_resolved),G.subgraph(SS1_resolved)


    

    
```

```python

def graphlist(graph_list):
    """
    this function inputs a list of n graphs and output a list of 2*n partitioned graphs.
    """ 
    new_graph_list = []
    for x in graph_list :
        a,b = graphpart(x)
        new_graph_list+=[a,b]
    return new_graph_list
        
```

```python

def graphsolve(G,k) :
    """
    this function takes as input a graph and outputs a list of subgraphs partitioned using Modularity multiple times.
    k is the number of partitions required which in this case needs to be a power of 2.
    """
    graph_list=[G]
    for i in range (k//2):
        temp= graphlist(graph_list)
        graph_list= temp
        
    return graph_list    
    
```

```python
#Cut Discrepancy

import numpy as np

def cut_discrepancy(true_labels, predicted_labels, distance_matrix):
    unique_true_labels, counts_true = np.unique(true_labels, return_counts=True)
    unique_predicted_labels, counts_predicted = np.unique(predicted_labels, return_counts=True)

    num_vertices = len(true_labels)
    
    # Create matrices for label differences and label counts
    true_label_diff = np.array([true_labels != label for label in unique_true_labels])
    predicted_label_diff = np.array([predicted_labels != label for label in unique_predicted_labels])

    # Calculate the cut size for true and predicted labels
    true_cut_size = np.sum(true_label_diff[:, :, None] * distance_matrix) / np.sqrt(np.outer(counts_true, counts_true))
    predicted_cut_size = np.sum(predicted_label_diff[:, :, None] * distance_matrix) / np.sqrt(np.outer(counts_predicted, counts_predicted))

    # Normalize the cut sizes by the total distance in the mesh
    true_normalized_cut = true_cut_size / np.sum(distance_matrix)
    predicted_normalized_cut = predicted_cut_size / np.sum(distance_matrix)

    # Calculate the cut discrepancy
    cut_discrepancy = np.abs(true_normalized_cut - predicted_normalized_cut)

    return cut_discrepancy

# Example usage:
# cut_disc = cut_discrepancy(true_labels, predicted_labels, distance_matrix)
```

```python
# Example usage
# Assume you have true_labels, predicted_labels, and distance_matrix
# true_labels: N-dimensional array of true labels defining the segments
# predicted_labels: N-dimensional array of predicted labels
# distance_matrix: NxN matrix of pairwise distances between nodes

# Calculate and print the cut discrepancy
cut_disc = cut_discrepancy(labels, labels_quantum, distance_matrix)
print(f"Cut Discrepancy: {cut_disc}")
```

```python
#Consistency Error

import numpy as np
from itertools import combinations

def consistency_error(partition1, partition2):
    """
    Calculate Consistency Error between two partitions.

    Parameters:
    - partition1: List or array-like, representing the first partition.
    - partition2: List or array-like, representing the second partition.

    Returns:
    - consistency_error: Consistency Error value.
    """

    def calculate_consistency(cluster1, cluster2):
        """
        Helper function to calculate consistency between two clusters.
        """
        common_pairs = sum(1 for pair in combinations(cluster1, 2) if pair[0] in cluster2 and pair[1] in cluster2)
        total_pairs = len(list(combinations(cluster1, 2)))
        return common_pairs / total_pairs if total_pairs > 0 else 1.0

    unique_labels1 = np.unique(partition1)
    unique_labels2 = np.unique(partition2)

    consistency_sum = 0.0
    total_combinations = 0

    for label1 in unique_labels1:
        for label2 in unique_labels2:
            if label1 != label2:
                cluster1 = np.where(partition1 == label1)[0]
                cluster2 = np.where(partition2 == label2)[0]

                consistency_sum += calculate_consistency(cluster1, cluster2)
                total_combinations += 1

    consistency_error = 1.0 - (consistency_sum / total_combinations) if total_combinations > 0 else 0.0

    return consistency_error


```

```python
# Example usage:
consistency_err = consistency_error(labels, labels_quantum)
```
